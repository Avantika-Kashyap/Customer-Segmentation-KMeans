```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. File ka naam 'Mall_Customers.csv' ekdum sahi kar diya hai
df = pd.read_csv('/content/Mall_Customers.csv')

# 2. Dataset se columns select karna
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

# 3. Khali WCSS list banana
wcss = []

# 4. K-Means Loop chalana
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# 5. Red rang ka Elbow Graph plot karna
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o', color='red', linewidth=2)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# 1. K-Means Model ko 5 clusters ke sath train karna
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# 2. Final Colorful Scatter Plot banana
plt.figure(figsize=(10, 6))

# Saare 5 clusters ke indexes (0, 1, 2, 3, 4) ekdum sahi kar diye hain:
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='pink', label='Cluster 1 (Careful)')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='yellow', label='Cluster 2 (Standard)')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='lightgreen', label='Cluster 3 (Target)') # <-- Line 13 ekdum sahi!
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='orange', label='Cluster 4 (Spendthrifts)')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='purple', label='Cluster 5 (Sensible)')

# Sabhi clusters ke beech ka main center point (Centroids) dikhana
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', marker='*', label='Centroids')

plt.title('Clusters of Customers (Final Output)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```
